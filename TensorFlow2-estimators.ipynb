# Using Estimators in TensorFlow 2.0

`tf.estimator` is a high-level TensorFlow API. Estimators encapsulate the following actions:
- training
- evaluation
- prediction
- export for serving

We may either use the pre-made Estimators provided or write our own custom [Estimators](https://www.tensorflow.org/guide/estimator). 

```python
# Import necessary modules
import pandas as pd
import numpy as np

import tensorflow as tf
from tensorflow import keras
```

```python
# Read data
df = pd.read_csv('data/tf/house.csv')
df.head()
```

I'll first define a list of feature columns, `feature_list`, and a data input function, `input_fn()`.

```python
# Define feature columns for bedrooms and bathrooms
bedrooms = tf.feature_column.numeric_column("bedrooms")
bathrooms = tf.feature_column.numeric_column("bathrooms")

# Define the list of feature columns
feature_list = [bedrooms, bathrooms]

def input_fn():
    # Define the labels
    labels = np.array(df.price)
    # Define the features
    features = {'bedrooms':np.array(df['bedrooms']), 
                'bathrooms':np.array(df['bathrooms'])}
    return features, labels
```

I'll now use the feature columns and data input function to define and train an estimator.

```python
# Define the model and set the number of steps
model = tf.estimator.DNNRegressor(feature_columns=feature_list, hidden_units=[2,2])
model.train(input_fn, steps=1)
```

```python
# Define the model and set the number of steps
model = tf.estimator.LinearRegressor(feature_columns=feature_list)
model.train(input_fn, steps=2)
```

Note that we have other premade estimator options, such as `BoostedTreesRegressor()`, and one can also create his/her own custom estimators.
